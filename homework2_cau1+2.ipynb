```
import pandas as pd
from collections import defaultdict

name=[]
region=[]
dict_plant=defaultdict(list)
with open('plants.data', encoding='ISO-8859-1') as f:
    lines=f.readlines()
    
for line in lines:
    line=line.strip('\n')
    if line.split(",")[0] not in name:
      name.append(line.split(",")[0])

    region_split=line.split(",")[1:]
    dict_plant[name[-1]]=region_split
    for r in region_split:
        if r not in region:
            region.append(r)
```

```
df=pd.DataFrame(columns=['names']+region)
for key, values in df.iteritems():
  df[key]=[0] * len(name)
df['names']=name

df
```

```
i=1
for key, values in dict_plant.items():
  for value in values:
    df.loc[df['names'] == key, value] = 1 
  print("----processed " +str(i)+" samples---\n")
  i+=1

df.to_csv('plants.csv', sep=',',index=None)
```

```
import numpy as np
freq=[]
for key, values in df.iteritems():
  count_plant_per_region = df.pivot_table(index=[key], aggfunc='size')
  print(key + " : "+str(count_plant_per_region[1]))
  freq.append(count_plant_per_region[1])

cal=np.array(freq)
cal=np.delete(cal,0)
```

```
#find min
cal.min()

#find max
cal.max()

#sum
cal.sum()
```
